```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "Resources/Mouse_metadata.csv"
study_results_path = "Resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_df = pd.merge(mouse_metadata, study_results, how = 'outer', on='Mouse ID')

# Display the data table for preview
merged_df.tail(5)
```

```python
# Checking the number of mice.
Num_Mice = len(merged_df["Mouse ID"].unique())
Num_Mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
Duplicates = merged_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)
Duplicated_Mice = merged_df.loc[Duplicates, 'Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
Duplicated_Data_df = merged_df.loc[merged_df['Mouse ID']==Duplicated_Mice[0]]

Duplicated_Data_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

Clean_df = pd.concat([Duplicated_Data_df, merged_df]).drop_duplicates(keep=False)
Clean_df.sample(10)
```

```python
# Checking the number of mice in the clean DataFrame.
Num_Mice_Drop = len(Clean_df["Mouse ID"].unique())
Num_Mice_Drop
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_tumor_mean = Clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
drug_tumor_median = Clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
drug_tumor_var = Clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
drug_tumor_std = Clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
drug_tumor_sems = Clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
drug_regimen = Clean_df['Drug Regimen'].unique()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

Summary_Statistics = {
    "Mean Tumor Volume (mm3)": drug_tumor_mean,
    "Median Tumor Volume (mm3)": drug_tumor_median,
    "Variance Tumor Volume (mm3)": drug_tumor_var,
    "Std Tumor Volume (mm3)": drug_tumor_std,
    "SEM Tumor Volume (mm3)": drug_tumor_sems
}
Summary_Statistics = pd.DataFrame(Summary_Statistics)
Summary_Statistics
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
concise_statistics = Clean_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
# Using the aggregation method, produce the same summary statistics in a single line
concise_statistics.columns = pd.MultiIndex.from_product([['Tumor Volume (mm3)'], concise_statistics.columns])
concise_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

specific_regimen_df = Clean_df[Clean_df['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]
specific_regimen_df.reset_index(drop=True, inplace=True)
 #Start by getting the last (greatest) timepoint for each mouse

last_timepoint_group = specific_regimen_df.groupby('Mouse ID').agg(
    Last_Timepoint=('Timepoint', 'last'),
    Last_Tumor_Volume=('Tumor Volume (mm3)', 'last')
).reset_index()


 #Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_timepoint_merge = pd.merge(Clean_df, last_timepoint_group, on='Mouse ID')
last_timepoint_merge

```

```python
Treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)

Tumor_Vol = []
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for i in Treatments:
    Tumor_df = last_timepoint_merge.loc[last_timepoint_merge['Drug Regimen'] == i]
    Tumor_vol_df = Tumor_df.loc[Tumor_df['Timepoint']==Tumor_df['Last_Timepoint']]
    volume = Tumor_vol_df['Tumor Volume (mm3)']
    volume.reset_index(drop=True, inplace=True)
    Tumor_Vol.append(volume)
    # Locate the rows which contain mice on each drug and get the tumor volumes
    IQR_quarts = volume.quantile([0.25,0.50,0.75])
    lower = IQR_quarts[0.25]
    upper = IQR_quarts[0.75]
    IQR = upper-lower
    lower_bounds = lower - (1.5*IQR)
    upper_bounds = upper + (1.5*IQR)
    # add subset 
    potential_outlier = (volume.loc[(Tumor_vol_df['Tumor Volume (mm3)'] >= upper_bounds) | (Tumor_vol_df['Tumor Volume (mm3)'] <= lower_bounds)])
    
    # Determine outliers using upper and lower bounds
    print(f'{i}s potential outliers: {potential_outlier}')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.boxplot(Tumor_Vol, labels = Treatments, sym='ro')
ax1.set_ylabel('Final Tumor Volume (mmm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulin = last_timepoint_merge.loc[last_timepoint_merge['Drug Regimen'] == 'Capomulin']
Capomulin_mouse = Capomulin.iloc[0,0]

combine_mouse_df = Capomulin.loc[Capomulin['Mouse ID'] == Capomulin_mouse]
x_axis = combine_mouse_df['Timepoint']
y_axis = combine_mouse_df['Tumor Volume (mm3)']
plt.title(f"Capomulin treatment of mouse {Capomulin_mouse}")
plt.xlabel("Timepoint(days)")
plt.ylabel("Tumor Volume (mm3)")

plt.plot(x_axis, y_axis)
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
average_tumor_size = Capomulin.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
average_size_df = pd.merge(Capomulin, average_tumor_size, on='Mouse ID')
average_size_df.plot(kind='scatter', x='Weight (g)', y = 'Tumor Volume (mm3)_y')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Capomulin Mouse Weight vs Average Observed Tumor Size')
plt.grid()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

average_size_df_sort = average_size_df[['Mouse ID', 'Weight (g)', "Tumor Volume (mm3)_y"]]
Correlation_df = average_size_df_sort.set_index('Mouse ID')
Correlation_df.corr()
Correlation = round(Correlation_df.corr().iloc[0,1], 2)
regression = st.linregress(average_size_df['Weight (g)'], average_size_df['Tumor Volume (mm3)_y'])
slope = regression[0]
intercept = regression[1]

regressed_values = average_size_df['Weight (g)']*slope+intercept
regressed_values

print(f'The correlation between mouse weight and the average tumor volume is {Correlation}')
plt.scatter(average_size_df['Weight (g)'], average_size_df['Tumor Volume (mm3)_y'])
plt.plot(average_size_df['Weight (g)'], regressed_values, color='r', alpha=0.7)
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Weight (g)')
plt.title(f'Capomulin Mouse Weight vs Average Observed Tumor Size')
plt.grid()
```
